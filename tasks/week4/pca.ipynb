```python
import pandas as pd
import numpy as np

close_prices = pd.read_csv('data/close_prices.csv')
```

```python
from sklearn.decomposition import PCA
train = close_prices.iloc[:, 1:]
pca = PCA(n_components=10)
pca_price = pca.fit_transform(train)
```

```python
for i in range(0, 10):
    if np.sum(pca.explained_variance_ratio_[:i]) > 0.9:
        np.savetxt('output/week4/task2_1.txt', [i], newline=' ', fmt='%i')
        break
```

```python
first_comp = pca_price[:, 0]
dj_ind = pd.read_csv('data/djia_index.csv')
corr = np.corrcoef(first_comp, dj_ind['^DJI'])[0, 1]
np.savetxt('output/week4/task2_2.txt', [np.round(corr, 2)], newline=' ', fmt='%.02f')
```

```python
best_ind = np.argmax(pca.components_[0])
np.savetxt('output/week4/task2_3.txt', [list(train)[best_ind]], newline='', fmt='%s')
```
